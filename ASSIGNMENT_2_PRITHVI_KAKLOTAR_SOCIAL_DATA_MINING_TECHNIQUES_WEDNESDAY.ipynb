```python
import pandas as pd
pd.set_option('display.max_colwidth',100)
data =pd.read_csv("ac.csv")
data.columns=['number','cols']
data
```

```python
copy=data.copy()
copy
```

```python
import string
string.punctuation
```

```python
k='!"#$%&\'()*+,.-/:;<=>?@[\\]^_`{|}~'
def remove_punct(txt):
    txt_nopunc =  "".join([char for char in txt if char not in k])
    return txt_nopunc
copy['cols_no_punc']=copy['cols'].apply(lambda x: remove_punct(x))
copy.head()
```

```python
import re
no_list=r'[0-9]'
copy['cols_no_digits']=copy['cols_no_punc'].apply(lambda x : re.sub(no_list,'',x.lower()))
copy.head()
```

```python
import nltk
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python

def remove_stopwords(txt):
    txt_clean=" ".join([word for word in (txt).split() if word not in stopwords])
    return txt_clean
copy['cols_no_stopwords']=copy['cols_no_digits'].apply(lambda x:remove_stopwords(x))
copy.head()
```

```python
import nltk
#Lemmatizing
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
wn = WordNetLemmatizer()
def lemmatization(txt):
    return " ".join([wn.lemmatize(word) for word in txt.split()])
copy['cols_lemm']=copy['cols_no_stopwords'].apply(lambda x:lemmatization(x))
copy.head()
```

```python
import nltk
#Lemmatizing
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
wn = WordNetLemmatizer()
def lemmatization(txt):
    text=[wn.lemmatize(word) for word in txt.split()]
    return text
copy['cols_lemm_word_count']=copy['cols_no_stopwords'].apply(lambda x:lemmatization(x))
copy.head()
```

```python
import nltk
nltk.download('punkt')
from nltk import word_tokenize
from nltk import sent_tokenize
from nltk.util import ngrams
list_bi=[]
for line in copy['cols_lemm']:
    token = word_tokenize(line)
    bigram = list(ngrams(token, 2))
    for i in bigram:
        list_bi.append(i)
print(list_bi)
    
```

```python
import collections
count=collections.Counter(list_bi).most_common()[:10]
print(count)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.bar( [str(i[0]) for i in count] ,[int(i[1]) for i in count] , color='yellow')
plt.xlabel("most common bigrams")
plt.ylabel(" count ")
plt.show()
```

```python
import nltk
nltk.download('punkt')
from nltk import word_tokenize
from nltk import sent_tokenize
from nltk.util import ngrams
list_tri=[]
for line in copy['cols_lemm']:
    token = word_tokenize(line)
    trigrams = list(ngrams(token, 3))
    for i in trigrams:
        list_tri.append(i)
print(list_tri)
```

```python
import collections
cnts=collections.Counter(list_tri).most_common()[:10]
print(cnts)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.bar( [str(i[0]) for i in cnts] ,[int(i[1]) for i in cnts] , color='yellow')
plt.xlabel("most common trigrams")
plt.ylabel(" count ")
plt.show()
```

```python
list_of_words=[]
for i in range(0,1648):
    y=copy['cols_lemm_word_count'].loc[i]
    for j in y:
        list_of_words.append(j)
total_words=len(list_of_words)
total_words
```

```python
unique_words=[]
for j in list_of_words:
    if j not in unique_words:
        unique_words.append(j)
cnt=len(unique_words)
cnt
```

```python
len_of_words=[]
for i in unique_words:
    y=len(i)
    len_of_words.append(y)
display(len_of_words)
```

```python
zip_iterator=zip(unique_words,len_of_words)
word_dict=dict(zip_iterator)
display(word_dict)
```

```python
import matplotlib.pyplot as plt
plt.hist(len_of_words,width=5,color='g',edgecolor='black', linewidth=1.2)
plt.show()
```

```python
import nltk
from nltk.tokenize import sent_tokenize
def sent(txt):
    text=[nltk.sent_tokenize(txt)]
    return text
copy['cols_sent']=copy['cols'].apply(lambda x: sent(x))
copy.head()
```

```python
list_sent=[]
final_list_sent=[]
for i in range(0,1648):
    s=copy['cols_sent'].loc[i]
    for j in s:
        list_sent.append(j)
for ii in range(len(list_sent)):
    for jj in list_sent[ii]:
        final_list_sent.append(jj)
```

```python
len_of_sent=[]
for k in final_list_sent:
    n=len(k)
    len_of_sent.append(n)
print(len_of_sent)
```

```python
zip_iterator=zip(final_list_sent,len_of_sent)
sent_dict=dict(zip_iterator)
display(sent_dict)
```
